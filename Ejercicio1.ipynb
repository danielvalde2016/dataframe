```python
from faker import Faker
import pandas as pd
import array as arr
from random import randint

fake = Faker()

aleatorio = randint(25, 50)
lista = ""

cols = ['nombre', 'direccion', 'telefono', 'email']
df = pd.DataFrame(columns=cols, index=range(aleatorio))
for a in range(aleatorio):
    nombre = fake.name()
    direccion = fake.address()
    telefono = fake.random_number(9)
    
    nick = fake.user_name()
    correo = '@'
    dominio = fake.free_email_domain()
    mail = nick + correo + dominio
    
    lista += "" + dominio + " "
    
    df.loc[a].nombre = nombre
    df.loc[a].direccion = direccion
    df.loc[a].telefono = telefono
    df.loc[a].email = mail
df



```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

cadenaPalabras = lista

listaPalabras = cadenaPalabras.split()

frecuenciaPalab = []
for w in listaPalabras:
    frecuenciaPalab.append(listaPalabras.count(w))


correos = ("Hotmail", "Gmail", "Yahoo")
posicion_y = np.arange(len(correos))
unidades = (pd.Series.value_counts(listaPalabras))
plt.barh(posicion_y, unidades, align = "center")
plt.yticks(posicion_y, correos)
plt.xlabel('Número de usuarios')
plt.title("Correos más utilizados")

```
